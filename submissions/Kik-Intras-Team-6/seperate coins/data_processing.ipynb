```python
import pandas as pd
import numpy as np
import glob
```

```python
# Load all coins
data = []
for file in glob.glob("../data/coin_*.csv"):
    df = pd.read_csv(file)
    df["Date"] = pd.to_datetime(df["Date"])
    df.name = file.replace("../data\coin_", "").replace(".csv", "")
    data.append(df)
    print(df.name)
```

```python
results = []
for group in data:
    g = group.sort_values("Date").copy()
    g["ret_1d"]     = g["Close"].pct_change(1)
    g["ret_7d"]     = g["Close"].pct_change(7)
    g["ma_7"]       = g["Close"].rolling(7).mean()
    g["ma_30"]      = g["Close"].rolling(30).mean()
    g["ma_ratio"]   = g["ma_7"] / g["ma_30"]
    g["volatility"] = g["ret_1d"].rolling(7).std()
    g["target"]     = g["Close"].pct_change(1)
    g["direction"] = (g["target"] > 0).astype(int).shift(1)
    g["day_of_week"] = g["Date"].dt.dayofweek   # 0=Monday, 6=Sunday
    g["month"]       = g["Date"].dt.month
    g["Volume"] = g["Volume"]
    g = g.drop(columns = ["Date", "Open" , "High", "Low", "Close" ])
    g.name = group.name
    results.append(g)
```
