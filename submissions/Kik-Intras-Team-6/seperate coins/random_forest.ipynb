```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Processed_Data/Bitcoin.csv").dropna()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
data = df.drop(columns=["direction", "month", "day_of_week"])
data.columns
```

```python
targets = df["direction"]
```

```python
X_train,X_test,y_train,y_test= train_test_split(data,targets,test_size=0.2, random_state=42) 
```

```python
X_test.isnull().sum()
```

```python
model = RandomForestClassifier(n_estimators=100,min_samples_split=200, max_depth=10,max_features='sqrt', random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy score: {accuracy:.10f}")
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# Get feature importances and names
importances = model.feature_importances_
feature_names = model.feature_names_in_

# Sort features by importance
indices = np.argsort(importances)
sorted_importances = importances[indices]
sorted_features = feature_names[indices]

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_features, sorted_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Random Forest Feature Importances (Stellar)')
plt.tight_layout()
plt.show()
```

```python
param_grid = {
    'n_estimators': list(range(50, 500, 50))+list(range(10, 40, 10)),
    'max_features': ['log2', 'sqrt'],
    'max_depth': range(2, 15, 2),
    'min_samples_split': range(50, 400, 75),
    'criterion': ['gini', 'entropy']
}
```

```python
rf = RandomForestClassifier(n_jobs=-1)
```

```python
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv = StratifiedKFold(3), n_jobs=-1, verbose=3)
```

```python
y_train.isnull().sum()
```

```python
grid_search.fit(X_train, y_train)
```

```python
y_pred = grid_search.predict(X_test)
y_pred_proba = grid_search.predict_proba(X_test)
```

```python
accuracy = accuracy_score(y_pred, y_test)
print(accuracy)
```

```python
import matplotlib.pyplot as plt

# Get feature importances and names
importances = grid_search.best_estimator_.feature_importances_
feature_names = grid_search.best_estimator_.feature_names_in_

# Sort features by importance
indices = np.argsort(importances)
sorted_importances = importances[indices]
sorted_features = feature_names[indices]

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_features, sorted_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Random Forest Feature Importances (Stellar)')
plt.tight_layout()
plt.show()
```
